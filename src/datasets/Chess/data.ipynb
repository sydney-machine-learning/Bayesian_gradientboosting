```python
import numpy as np
import pandas as pd
```

```python
pd.read_csv('chess.csv', header=None, delimiter=';')
```

```python
from sklearn.model_selection import train_test_split
```

```python
data = np.genfromtxt('chess.csv', delimiter=';')
```

```python
X = data[:, :6]
y = data[:, 6:]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
data_train = np.concatenate([X_train, y_train], axis=1)
data_test = np.concatenate([X_test, y_test], axis=1)
```

```python
np.savetxt('train.csv', data_train, delimiter=',')
np.savetxt('test.csv', data_test, delimiter=',')
```
